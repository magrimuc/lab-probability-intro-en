## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
import math
# Ball counts
white = 10
red = 12
black = 8
total = white + red + black # 30

# Probability of drawing a white ball
p_white = white / total
print (f"Propability of white: {p_white}")

firstwhiteThenblack = p_white * black / (total - 1)
print (f"Propability of first white then black: {firstwhiteThenblack}")

firstwhiteThenblackwithreplacement = p_white * black / total
print (f"Propability of first white then black with replacement: {firstwhiteThenblackwithreplacement}")

tot = math.comb(total, 3) # $\binom{30}{3} = 4060$
print (f"All possible combinations: {tot}")

fav = math.comb(white, 1) * math.comb(red, 1) * math.comb(black, 1) # $\binom{10}{1} \cdot \binom{12}{1} \cdot \binom{8}{1} = 960$
print (f"Favorable combinations: {fav}")

p_fav = fav / tot
print (f"Favorable combinations / all possible combinations: {p_fav}")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
cl = 0.4
rifcl = 0.5 # rains if cloudy
rifncl = 0.2 # rains if not cloudy

prop = cl * rifcl + (1 - cl) * rifncl # prop of being cloudy and it rains + it's not cloudy and it rains anyway
print (f"propability of rain in both cases: cloudy and not cloudy: {prop}")
```

a day may be cloudy then 0.4 * 0.5 adds up to rainy if cloudy prob.

a day may be non-cloudy then 0.6 * 0.2 adds up to rainy if not cloudy prob.

We sum up both probs and get 0.32 chance of rain next Saturday.
